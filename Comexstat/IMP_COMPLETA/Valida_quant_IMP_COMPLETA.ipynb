```python
from sklearn.preprocessing import MinMaxScaler as MMS
MMS=MMS()
sc=pd.DataFrame(MMS.fit_transform(imp_fert_bim[['log_kg','log_fob','fob_kg','frete_kg']]), columns=['log_kg_s','log_fob_s','fob_kg_s','frete_kg_s'])
sc.head(1)
```

```python
imp_fert_bim=pd.concat([imp_fert_bim.reset_index(),sc],axis=1)
```

```python
q=pd.DataFrame()
from sklearn.cluster import KMeans as KM
KM=KM(n_clusters=2, n_init='auto')
for ncm in imp_fert_bim.CO_NCM.unique():
    imp_fert_bim_ncm=imp_fert_bim[imp_fert_bim.CO_NCM==ncm]
    imp_fert_bim_ncm=imp_fert_bim_ncm.assign(km=KM.fit_predict(imp_fert_bim_ncm[['log_kg_s','log_fob_s','fob_kg_s']]))
    q=pd.concat([q,imp_fert_bim_ncm])
```

```python
# Cria df somente com os últimos 5 anos
imp_fert_5=imp_fert[imp_fert.CO_ANO>imp_fert.CO_ANO.max()-5]
```

```python
#Para os últimos 5 anos, cria df com rank dos NCM por KG, em seguida selecioina somente os NCMs maiores, até que eles cubram 80% do peso total de importado
imp_fert_perct=(imp_fert_5.groupby('CO_NCM').KG_LIQUIDO.sum()/imp_fert_5.KG_LIQUIDO.sum()*100).sort_values(ascending=False).astype(int).reset_index().rename(columns={'KG_LIQUIDO':'total_perct'})
imp_fert_perct.head()
```

```python
# Elimina alerta de versão
pd.set_option('future.no_silent_downcasting', True)
```

```python
# Cria coluna com valores da percetagem em peso acumulada do maior para o menor.
imp_fert_perct['tp_acml'] = imp_fert_perct['total_perct'].rolling(window=len(imp_fert_perct), min_periods=1).sum()
imp_fert_perct
# Cria coluna para marcar True, do topo para baixo, quais as percentagens acumuladas são maiores que 80%.
imp_fert_perct.loc[imp_fert_perct.tp_acml>80,'above_80']=True
imp_fert_perct
# Na mesma coluna, marca False para as demais
imp_fert_perct=imp_fert_perct.fillna(False).copy()

imp_fert_perct['ab_acml'] = imp_fert_perct['above_80'].rolling(window=len(imp_fert_perct), min_periods=1).sum()

imp_fert_perct_top=imp_fert_perct[imp_fert_perct.ab_acml<=1][['CO_NCM']]

print(imp_fert_perct_top)
```

```python
imp_fert=imp_fert.merge(imp_fert_perct_top)
```
