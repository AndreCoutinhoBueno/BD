```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Carrega a tabela imortação
imp_fert=pd.read_csv('~/git/BD/temp/imp_fert1.csv').reset_index(drop=True)
```

# CO_ANO e CO_MES

```python
# Lista os anos
lista_anos = imp_fert.CO_ANO.unique().tolist()

def verificar_anos(anos):
    # Ordenar a lista de anos
    anos_ordenados = sorted(anos)
    
    # Verificar se todos os anos no intervalo estão presentes na lista
    for ano in range(anos_ordenados[0], anos_ordenados[-1] + 1):
        if ano not in anos_ordenados:
            return False
    return True

if verificar_anos(lista_anos):
    print(f"A tabela apresenta todos os anos entre {imp_fert.CO_ANO.min()} e {imp_fert.CO_ANO.max()}.")
else:
    print("A tabela não apresenta continuidade entre os anos.")


def verificar_tabela_completa(tabela):
    # Gerar todos os pares possíveis de CO_ANO e mês no intervalo da tabela
    CO_ANOs = range(tabela['CO_ANO'].min(), tabela['CO_ANO'].max() + 1)
    
    # Encontrar o último CO_ANO e o último mês registrado
    ultimo_CO_ANO = tabela['CO_ANO'].max()
    ultimo_CO_MES = tabela[tabela['CO_ANO'] == ultimo_CO_ANO]['CO_MES'].max()
    
    todos_pares = set()
    for CO_ANO in CO_ANOs:
        if CO_ANO == ultimo_CO_ANO:
            CO_MESes = range(1, ultimo_CO_MES + 1)
        else:
            CO_MESes = range(1, 13)
        todos_pares.update((CO_ANO, CO_MES) for CO_MES in CO_MESes)
    
    # Gerar os pares presentes na tabela
    pares_presentes = set(zip(tabela['CO_ANO'], tabela['CO_MES']))
    
    # Verificar se todos os pares possíveis estão presentes na tabela
    pares_faltando = todos_pares - pares_presentes
    
    if pares_faltando:
        print("A tabela não apresenta continuidade nos seguintes momentos:")
        for par in sorted(pares_faltando):
            print(par)
        #return False
    else:
        print("A tabela apresenta continuidade entre os meses.")
        #return True

verificar_tabela_completa(imp_fert)
```

# Normalidade

Antes de utilizar  valores quantitativos para ranquear e filtrar categorias, será conferida a normalidade geral:

```python
sns.displot(height=3,data=imp_fert,x='KG_LIQUIDO',log_scale=True)
```

```python
imp_fert=imp_fert[imp_fert.KG_LIQUIDO>0]
imp_fert=imp_fert.assign(log_kg=np.log10(imp_fert.KG_LIQUIDO))
```

```python
from sklearn.cluster import KMeans as KM
KM=KM(n_clusters=3)
imp_fert=imp_fert.assign(km=KM.fit_predict(imp_fert[['log_kg']]))

sns.displot(height=3,data=imp_fert,x='KG_LIQUIDO',log_scale=True,hue='km')
```

Serão marcados como anormais registros menores que 10^3 e maiores que 10^8.

```python
km_top=imp_fert.groupby('km').KG_LIQUIDO.mean().sort_values(ascending=False).reset_index().km.head(1).values[0]
```

```python
imp_fert=imp_fert.assign(anormal=0)

imp_fert.loc[(imp_fert.km!=km_top)&(imp_fert.KG_LIQUIDO<10**8),'anormal']=1
```

# CO_NCM

```python
# Ranqueia a seleciona os 10 maiores NCMs em peso
imp_fert_top=imp_fert[imp_fert.anormal==0].groupby('CO_NCM').KG_LIQUIDO.sum().sort_values(ascending=False).reset_index().head(10)['CO_NCM'].values

# Filtra a tabela deixano passar somente os registros dos NCMs top
imp_fert=imp_fert[imp_fert.CO_NCM.isin(imp_fert_top)]

# Carrega a tabela complementar NCM
ncm=pd.read_csv('~/git/BD/Comexstat/tab_complementares/NCM.csv', encoding='Latin-1',sep=';')

# Funde as tabelas
imp_fert=imp_fert.merge(ncm)

tab_name=imp_fert[['CO_NCM','NO_NCM_POR']].drop_duplicates().assign(sigla=imp_fert.NO_NCM_POR).sort_values('sigla')

pd.options.display.max_colwidth = 250

print(tab_name[['NO_NCM_POR']].drop_duplicates())
```

```python
# Constrói, com base na tabela impressa acima, uma lista de siglas para facilitar a identificação dos produtos
sigla=['NPK','MAP','NAM','NP','MOP','OSP','SAM','SSP','TSP','UREIA']

tab_name.sigla=sigla

imp_fert=imp_fert.merge(tab_name)

# Define o título dos gráficos
title='Importação de Fertilizantes'

# Insere coluna com a data
imp_fert.insert(0,'date',pd.to_datetime(imp_fert.CO_ANO.astype(str)+'/'+imp_fert.CO_MES.astype(str)+'/1'))

# Visualiza cronologicamenete os pesos por NCM
sns.relplot(data=imp_fert[imp_fert.anormal==0],x='date',y='KG_LIQUIDO', col='sigla', col_wrap=5, facet_kws=dict(sharey=False))\
.fig.suptitle(title,y=1.1)
plt.show()

# Visualiza cronologicamenete os valores por NCM
sns.relplot(data=imp_fert[imp_fert.anormal==0],x='date',y='VL_FOB', col='sigla', col_wrap=5, facet_kws=dict(sharey=False))\
.fig.suptitle(title,y=1.1)
plt.show()

# Descarta os 3 tipos de superfosfatos, pois estes produtos apresentam complexidade extra na composição química, que levaram a alteração do
# código NCM durante o período de registro.
descarte=['SSP','TSP','OSP','XSP']
imp_fert=imp_fert[~imp_fert.sigla.isin(descarte)]

print(f"Os NCMs selecionados para seguirem no projeto são:\n\n{imp_fert[['sigla','CO_NCM']].drop_duplicates().set_index('sigla')}")
```

# CO_PAIS

```python
print(f"Existem {len(imp_fert.CO_PAIS.unique())} países exportadores.\
 Vamos manter identificados os 10 maiores e os restantes serão denominados como 'Não Declarados' , recebendo o código 999.")

rank_paises=imp_fert[imp_fert.anormal==0].groupby('CO_PAIS').KG_LIQUIDO.sum().sort_values(ascending=False).reset_index()

top_10_paises=rank_paises['CO_PAIS'].head(10)

imp_fert.loc[~imp_fert.CO_PAIS.isin(top_10_paises),'CO_PAIS']=999
```

# SG_UF_NCM

```python
print(f"Existem {len(imp_fert.SG_UF_NCM.unique())} unidades da federação com imortações registradas.\n\
As dez maiores terão a identidade mantida e as demais passarão a ser identificadas como 'outras', recebendo o código 0.")

top_10_UF=imp_fert[imp_fert.anormal==0].groupby('SG_UF_NCM').KG_LIQUIDO.sum().sort_values(ascending=False).reset_index().SG_UF_NCM.head(10)

imp_fert.loc[~imp_fert.SG_UF_NCM.isin(top_10_UF),'SG_UF_NCM']=0
```

# CO_VIA

```python
# Identifica a via líder em movimentação 
via=pd.read_csv('~/git/BD/Comexstat/tab_complementares/VIA.csv',encoding='Latin-1',sep=';')

rank_via=imp_fert[imp_fert.anormal==0].groupby('CO_VIA').KG_LIQUIDO.sum().sort_values(ascending=False).reset_index()

rank_via=rank_via.merge(via)

print(f"A via {rank_via.head(1).NO_VIA[0]} contém {round(rank_via.KG_LIQUIDO.values[0]/imp_fert.KG_LIQUIDO.sum()*100,2)} % do peso total registrado.\n\
As importações por outras vias serão descartadas.")

imp_fert=imp_fert[imp_fert.CO_VIA==1]
```

# CO_URF

```python
# Identifica a via líder em movimentação 
urf=pd.read_csv('~/git/BD/Comexstat/tab_complementares/URF.csv',encoding='Latin-1',sep=';')

rank_urf=imp_fert[imp_fert.anormal==0].groupby('CO_URF').KG_LIQUIDO.sum().sort_values(ascending=False)

print(f"Existem {len(rank_urf)} unidades da receita federal com registros de importações.\n\n\
As 10 maiores terão sua identificação mantida e as demais passarão a ser ideniticadas como 'Outras', recebendo o código 9999999.")

top_10_urf=rank_urf.reset_index().CO_URF.head(10)

imp_fert.loc[~imp_fert.CO_URF.isin(top_10_urf),'CO_URF']=9999999
```
