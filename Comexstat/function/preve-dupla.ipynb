```python
def preve(comex_df_NCM,nick_list,date_pred,comp_min,comp_max,md_c,md_r,quanti_col):

    # Confere comprimento mínimo
    if pd.to_datetime(str(date_pred.year-comp_min)+'/'+str(date_pred.month)+'/1') >= comex_df_NCM.date.min():
        
        '''############### Limita o comprimento máximo ##############################################
        ad_y=0
        if comp_max!=None:
            month_ini=date_pred.month+1
            if month_ini>12:
                month_ini=1
                ad_y=1
            data_inicio=pd.to_datetime(str(date_pred.year-comp_max+ad_y)+'/'+str(month_ini)+'/1')
            comex_df_NCM=comex_df_NCM[comex_df_NCM.date>=data_inicio]
        '''

        
        ################## Produz a Matriz Esparsa marcando as ocorrências ###########
        %run '~/git/BD/Comexstat/function/matriz_esparsa.ipynb'
        comex_df_NCM=matriz_esparsa(comex_df_NCM, quanti_col)        
        ##############################################################################
        
        
        # Parametriza os algoritmos
        RFC=RFC_(max_depth=md_c)
        RFR=RFR_(max_depth=md_r)
        
        # Adequa medidas de tempo para os algoritmos
        comex_df_NCM.date=comex_df_NCM.date.astype(int)
        comex_df_NCM.drop(columns='CO_ANO',inplace=True)

        # Codifica categóricos
        imp_fert_dumm=pd.get_dummies(comex_df_NCM,prefix_sep='~')

        # Separa treino e teste
        train_max=imp_fert_dumm[imp_fert_dumm.date< imp_fert_dumm.date.max()]
        test_max =imp_fert_dumm[imp_fert_dumm.date==imp_fert_dumm.date.max()]
        
        train_min=imp_fert_dumm[imp_fert_dumm.date> imp_fert_dumm.date.min()]
        test_min =imp_fert_dumm[imp_fert_dumm.date==imp_fert_dumm.date.min()]
       
        

        # Inicia df para coletar resultados de classificação e regressão do loop
        q_max=pd.DataFrame()
        q_min=pd.DataFrame()
        
        r_max=pd.DataFrame()
        r_min=pd.DataFrame()

        # inicia loop de repetição (para teste de variação dos algorítmos randômicos)
        for x in range(rept):

            # Identifica X e y classifier
            X_train_max=train_max.drop(columns=[quanti_col,'oc'])
            y_train_max=train_max.oc
            X_test_max =test_max.drop(columns=[quanti_col,'oc'])
            y_test_max =test_max.oc
            

            X_train_min=train_min.drop(columns=[quanti_col,'oc'])
            y_train_min=train_min.oc
            X_test_min =test_min.drop(columns=[quanti_col,'oc'])
            y_test_min =test_min.oc

        
            # ******* CLASSIFCAÇÃO *******

            # MAX
            # Fit classifier
            RFC.fit(X_train_max,y_train_max)
            # Predict classifier
            test_pred_max=X_test_max.assign(oc_pred=RFC.predict(X_test_max)).assign(oc=y_test_max)\
            .assign(nick_list=str(nick_list))\
            .assign(date_pred=date_pred)\
            .assign(rept=x)
            # Armazena
            q_max=pd.concat([q_max,test_pred_max])
            # MIN
            # Fit classifier
            RFC.fit(X_train_min,y_train_min)
            # Predict classifier
            test_pred_min=X_test_min.assign(oc_pred=RFC.predict(X_test_min)).assign(oc=y_test_min)\
            .assign(nick_list=str(nick_list))\
            .assign(date_pred=date_pred)\
            .assign(rept=x)
            # Armazena
            q_min=pd.concat([q_min,test_pred_min])

            
            # ******* REGRESSÃO *******
            
            # Filtra deixando passar registros ocorridos
            train_max=train_max[train_max.oc==1]
            test_max=test_max[test_max.oc==1]
            train_min=train_min[train_min.oc==1]
            test_min=test_min[test_min.oc==1]
            
            
            # verifica o comprimento mínimo
            if (len(train_max)>0)&(len(test_max)>0):
                
                # Identifica X e y
                X_train_max=train_max.drop(columns=[quanti_col,'oc'])
                y_train_max=train_max.KG_LIQUIDO
                X_test_max =test_max.drop(columns=[quanti_col,'oc'])
                y_test_max =test_max.KG_LIQUIDO
                
                X_train_min=train_min.drop(columns=[quanti_col,'oc'])
                y_train_min=train_min.KG_LIQUIDO
                X_test_min =test_min.drop(columns=[quanti_col,'oc'])
                y_test_min =test_min.KG_LIQUIDO
                
                # Fit regressor
                RFR.fit(X_train_max,y_train_max)
                # Predict classifier
                test_pred_max=X_test_max.assign(kg_pred=RFR.predict(X_test_max)).assign(KG_LIQUIDO=y_test_max)\
                .assign(nick_list=str(nick_list))\
                .assign(date_pred=date_pred)\
                .assign(rept=x)\
                .assign(prev='max')
                # Armazena
                r_max=pd.concat([r_max,test_pred_max])
                
                RFR.fit(X_train_min,y_train_min)
                # Predict classifier
                test_pred_min=X_test_min.assign(kg_pred=RFR.predict(X_test_min)).assign(KG_LIQUIDO=y_test_min)\
                .assign(nick_list=str(nick_list))\
                .assign(date_pred=date_pred)\
                .assign(rept=x)\
                .assign(prev='min')
                # Armazena
                r_min=pd.concat([r_min,test_pred_min])
    
        def f_dumm(df_dumm):
            '''Descodifica colunas categóricas'''
            a=pd.from_dummies(df_dumm.select_dtypes(include='bool'),sep='~')
            b=df_dumm.select_dtypes(exclude='bool')
            c=pd.concat([a,b],axis=1)
            return c
            
        # Descodifica colunas categóricas
        q_max=f_dumm(q_max)
        q_min=f_dumm(q_min)
        r_max=f_dumm(r_max)        
        r_min=f_dumm(r_min)
        
        
        score=q_max.merge(r_max,how='outer').merge(q_min,how='outer').merge(r_min,how='outer')
    else:
        score=pd.DataFrame()
    return score
```
