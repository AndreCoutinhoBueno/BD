### Clusterização

Vamos separar ou 'clusterizar' os NCMs bimodais em dois grupos, passando eles a unimodais.

Como a normalidade se apresenta em escala logarítmica, vamos proceder a 'clusterização' também nessa escala. Para isso, novas colunas comos  valores logarítmicos serão criadas.

```python
exp_agr=exp_agr.assign(log_kg=np.log10(exp_agr.KG_LIQUIDO))
exp_agr=exp_agr.assign(log_fob=np.log10(exp_agr.VL_FOB))
exp_agr=exp_agr.assign(fob_kg=exp_agr.VL_FOB/exp_agr.KG_LIQUIDO)
```

Além da escala logarítmica, os valores serão reescalonados para uma mesma escala entre eles, de 0 a 1, como é necessário ao bom funcionamento do algoritmo utilizado.

```python
from sklearn.preprocessing import MinMaxScaler as MMS
MMS=MMS()
sc=pd.DataFrame(MMS.fit_transform(exp_agr[['log_kg','log_fob','fob_kg']]), columns=['log_kg_s','log_fob_s','fob_kg_s'])

exp_agr=pd.concat([exp_agr.reset_index(drop=True),sc],axis=1)
```

```python
# Clusteriza pelo preço
q=pd.DataFrame()
from sklearn.cluster import KMeans as KM
KM=KM(n_clusters=9, n_init='auto', random_state=1)
for ncm in exp_agr.CO_NCM.unique():
    exp_agr_ncm=exp_agr[exp_agr.CO_NCM==ncm]
    exp_agr_km_ncm=exp_agr_ncm.assign(km=KM.fit_predict(exp_agr_ncm[['fob_kg_s']]))
    q=pd.concat([q,exp_agr_km_ncm])

# retoma o nome da tabla
exp_agr_km=q
```

```python
sns.displot(height=1.6,aspect=3,data=exp_agr_km,x='log_kg_s', col='CO_NCM', hue='km', bins=20, facet_kws={'sharex':False, 'sharey':False})
```

```python
# Elimina a clusterização dos NCMs em que ela não resultou no maior cluster normalizado
exp_agr_km.loc[exp_agr_km.CO_NCM==9011110,'km']=0
exp_agr_km.loc[exp_agr_km.CO_NCM==12019000,'km']=0
```

```python
# Adciona novo dígito ao NCM dos produtos em nova coluna
exp_agr_km['CO_NCM_km']=exp_agr_km['CO_NCM']
exp_agr_km['CO_NCM_km']=exp_agr_km.CO_NCM.astype(str)+'-'+exp_agr_km.km.astype(str)
```

```python
# Determina características dos clusters para identificação deles posteriormente
q=pd.DataFrame()
for ncm in exp_agr_km.CO_NCM_km.unique():
    exp_agr_km_ncm=exp_agr_km[exp_agr_km.CO_NCM_km==ncm]
    exp_agr_km_ncm=exp_agr_km_ncm.assign(média=exp_agr_km_ncm.KG_LIQUIDO.mean())
    exp_agr_km_ncm=exp_agr_km_ncm.assign(std=exp_agr_km_ncm.KG_LIQUIDO.std())
    q=pd.concat([q,exp_agr_km_ncm])
exp_agr_km=q
exp_agr_km.head(1)
```

```python
# Constrói e salva nova tabela com descrição dos novos NCMs.
NCM=pd.read_csv('~/git/BD/Comexstat/tab_complementares/NCM.csv',encoding='Latin-1',sep=';').astype(str)


tab_ncm=exp_agr_km[['CO_NCM','CO_NCM_km','média','std']].drop_duplicates()
tab_ncm.reset_index(drop=True)

tab_ncm=NCM[['CO_NCM','CO_SH6','NO_NCM_POR']].merge(tab_ncm)

# Salva nova tabela de NCM com as modificações referentes aos NCM bimodais
tab_ncm.to_csv('~/git/BD/Comexstat/tab_complementares/NCM_exp_modificado.csv',index=False)
```
