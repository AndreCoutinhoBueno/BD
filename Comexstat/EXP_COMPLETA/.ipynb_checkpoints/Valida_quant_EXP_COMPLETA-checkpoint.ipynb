# NCM

Verificamos que existe ampla diferença da quantidade de registros existente para os diferentes NCMs. Deixando a situação estatística desproporcional entre eles.

Assim, a seguir vamos filtrar os NCMs deixando passar somente aqueles que apresentem mais do que 7% do número de registros do NCM de maior número de registros.

```python
imp_fert_ncm=imp_fert.groupby('CO_NCM').KG_LIQUIDO.count().sort_values(ascending=False).reset_index().rename(columns={'KG_LIQUIDO':'counter'})

imp_fert_ncm=imp_fert_ncm[imp_fert_ncm.counter>(imp_fert_ncm.counter.max()*0.07)][['CO_NCM']]

imp_fert=imp_fert_ncm.merge(imp_fert)

sns.displot(data=imp_fert,x='KG_LIQUIDO',log_scale=True,col='CO_NCM',col_wrap=5)
```

Verificamos que vários dos NCMs apresentam dois cumes ou, apresentam distribuição bimodal. 
Para tratar o problema da bimodalidade, a tabela de importação será dividida em duas:
* com os NCMs normais ou unimodais
* com os NCM bimodais.

```python
ncm_bimodal_list=[
   31021010,31022100,31029000 ,31042090,31052000,31054000,31055900
]
```

```python
imp_fert_bim=imp_fert[imp_fert.CO_NCM.isin(ncm_bimodal_list)]
imp_fert_uni=imp_fert[~imp_fert.CO_NCM.isin(ncm_bimodal_list)]
```

### Clusterização

Vamos separar ou 'clusterizar' os NCMs bimodais em dois grupos, passando eles a unimodais.

Como a normalidade se apresenta em escala logarítmica, vamos proceder a 'clusterização' também nessa escala. Para isso, novas colunas comos  valores logarítmicos serão criadas.

```python
imp_fert_bim=imp_fert_bim.assign(log_kg=np.log10(imp_fert_bim.KG_LIQUIDO))
imp_fert_bim=imp_fert_bim.assign(log_fob=np.log10(imp_fert_bim.VL_FOB))

imp_fert_bim=imp_fert_bim.assign(fob_kg=imp_fert_bim.VL_FOB/imp_fert_bim.KG_LIQUIDO)
imp_fert_bim=imp_fert_bim.assign(frete_kg=imp_fert_bim.VL_FRETE/imp_fert_bim.KG_LIQUIDO)

imp_fert_bim=imp_fert_bim.assign(log_fob_kg=imp_fert_bim.log_fob/imp_fert_bim.log_kg)
```

Além da escala logarítmica, os valores serão reescalonados para uma mesma escala entre eles, de 0 a 1, como é necessário ao bom funcionamento do algoritmo utilizado.

```python
from sklearn.preprocessing import MinMaxScaler as MMS
MMS=MMS()
sc=pd.DataFrame(MMS.fit_transform(imp_fert_bim[['log_kg','log_fob','fob_kg','frete_kg']]), columns=['log_kg_s','log_fob_s','fob_kg_s','frete_kg_s'])

imp_fert_bim=pd.concat([imp_fert_bim.reset_index(drop=True),sc],axis=1)
```

```python
# Clusteriza
q=pd.DataFrame()
from sklearn.cluster import KMeans as KM
KM=KM(n_clusters=2, n_init='auto')
for ncm in imp_fert_bim.CO_NCM.unique():
    imp_fert_bim_ncm=imp_fert_bim[imp_fert_bim.CO_NCM==ncm]
    imp_fert_bim_ncm=imp_fert_bim_ncm.assign(km=KM.fit_predict(imp_fert_bim_ncm[['log_kg_s','log_fob_s','fob_kg_s']]))
    q=pd.concat([q,imp_fert_bim_ncm])

# retoma o nome da tabla
imp_fert_bim_corr=q

sns.displot(data=imp_fert_bim_corr,x='KG_LIQUIDO',log_scale=True, col='CO_NCM',col_wrap=5, hue='km')
```

Cada produto bimodal deu origem a dois produtos unimodais, que agora tem mais um dígito adicionado ao NCM, para poder identificá-los um do outro.

```python
# Adciona novo dígito ao NCM dos produtos originalmente bimodais
imp_fert_bim_corr['CO_NCM_orig']=imp_fert_bim_corr['CO_NCM']
imp_fert_bim_corr['CO_NCM']=imp_fert_bim_corr.CO_NCM.astype(str)+'-'+imp_fert_bim_corr.km.astype(str)
```

Em consequência, é necessário ampliar a tabela complementar com os NCMs, subdividindo também nela os NCMs bimodais, abrindo novo campo para identificar as características diferencias dos novos NCMs criados.

```python
NCM=pd.read_csv('~/git/BD/Comexstat/tab_complementares/NCM.csv',encoding='Latin-1',sep=';')

q=pd.DataFrame()
for ncm in imp_fert_bim_corr.CO_NCM.unique():
    imp_fert_bim_corr_ncm=imp_fert_bim_corr[imp_fert_bim_corr.CO_NCM==ncm]
    imp_fert_bim_corr_ncm=imp_fert_bim_corr_ncm.assign(média=imp_fert_bim_corr_ncm.KG_LIQUIDO.mean())
    imp_fert_bim_corr_ncm=imp_fert_bim_corr_ncm.assign(std=imp_fert_bim_corr_ncm.KG_LIQUIDO.std())
    q=pd.concat([q,imp_fert_bim_corr_ncm])
imp_fert_bim_corr_ncm=q

set_sub=set(np.arange(11,21))

set_col=set(np.arange(24))

col_rest=list(set_col-set_sub)

imp_fert_bim_corr_ncm=imp_fert_bim_corr_ncm[imp_fert_bim_corr_ncm.columns[col_rest]]

recode_ncm=imp_fert_bim_corr_ncm[['CO_NCM','CO_NCM_orig','média','std']].drop_duplicates()

tab_ncm=NCM[['CO_NCM','CO_SH6','NO_NCM_POR']].merge(recode_ncm, left_on='CO_NCM', right_on='CO_NCM_orig')\
.drop(columns='CO_NCM_x').rename(columns={'CO_NCM_y':'CO_NCM'})

# Salva nova tabela de NCM com as modificações referentes aos NCM bimodais
tab_ncm.to_csv('~/git/BD/Comexstat/tab_complementares/NCM_modificado.csv',index=False)

tab_ncm.head()
```

Concatena as tabelas com NCM unimodal e a tabela  bimodal corrigida

```python
# Concatena os registros originalmente unimodais com os registros bimodais que foram corrigdos para unimodal
imp_fert=pd.concat([imp_fert_uni,imp_fert_bim_corr]).dropna(axis=1)
imp_fert.head(1)
```

```python
# Cria df somente com os últimos 5 anos
imp_fert_5=imp_fert[imp_fert.CO_ANO>imp_fert.CO_ANO.max()-5]

#Para os últimos 5 anos, cria novo df com rank dos NCMs por KG importado, em seguida selecioina somente os NCMs maiores,
# até que eles cubram 80% do peso total de importado neste período
imp_fert_perct=\
(imp_fert_5.groupby('CO_NCM').KG_LIQUIDO.sum()/imp_fert_5.KG_LIQUIDO.sum()*100).sort_values(ascending=False)\
.astype(int).reset_index().rename(columns={'KG_LIQUIDO':'total_perct'})

# Cria coluna com valores da percetagem em peso acumulada do maior para o menor.
imp_fert_perct['tp_acml'] = imp_fert_perct['total_perct'].rolling(window=len(imp_fert_perct), min_periods=1).sum()

# Cria coluna para marcar 1, do topo para baixo, quais as percentagens acumuladas são maiores que 80%.
imp_fert_perct.loc[imp_fert_perct.tp_acml>80,'above_80']=1

# Na mesma coluna, marca 0 para as demais
imp_fert_perct=imp_fert_perct.fillna(0).copy()

imp_fert_perct_top=imp_fert_perct[imp_fert_perct.above_80==0][['CO_NCM']].astype(str)

imp_fert=imp_fert.merge(imp_fert_perct_top)

imp_fert.head(1)
```
